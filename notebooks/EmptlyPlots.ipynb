```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x**3, label='serial')  # Plot some data on the axes.
ax.plot(x, x**2, label='data parapllel')  # Plot more data on the axes...
ax.plot(x, x, label='data parallel block distributed')  # ... and some more.
ax.set_xlabel('log size')  # Add an x-label to the axes.
ax.set_ylabel('runtime')  # Add a y-label to the axes.
ax.set_title("Runtime vs Log Size")  # Add a title to the axes.
ax.legend()  # Add a legend.
plt.savefig('runtime_vs_logsize.svg')  
```
